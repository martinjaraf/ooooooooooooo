````python
by_score = data_animes.sort_values('Score',ascending=False)
top_5 = by_score.head(5)
top_5
````

````python
name_top_5 = top_5['Name']
completed_top_5 = top_5['Completed']
watching_top_5 = top_5['Watching']
plantowatch_top_5 = top_5['Plan to Watch']
plantowatch_top_5
````

````python
pie_chart_completed = plt.figure(figsize=(10,6))
tasks = completed_top_5.tolist()

my_labels = name_top_5.tolist()
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('completados top 5 animes')
plt.axis('equal')
plt.show()
````

````python
pie_chart_watching = plt.figure(figsize=(10,6))
tasks = watching_top_5.tolist()

my_labels = name_top_5.tolist()
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('viendo de los top 5 animes')
plt.axis('equal')
plt.show()
````

````python
pie_chart_plantowatch = plt.figure(figsize=(10,6))
tasks = plantowatch_top_5.tolist()

my_labels = name_top_5.tolist()
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('planeados para ver top 5 animes')
plt.axis('equal')
plt.show()
````

````python
total_completed = completed_top_5.sum()
total_watching = watching_top_5.sum()
total_plantowatch = plantowatch_top_5.sum()
````

````python
pie_chart_totales = plt.figure(figsize=(10,6))
tasks = [total_completed, total_watching, total_plantowatch]

my_labels = ['Completed', 'Watching', 'Plan to Watch']
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('Anime agregated')
plt.axis('equal')
plt.show()
````

````python
data_animes['Year'] = pd.to_numeric(data_animes['Year'], errors='coerce')
dat = data_animes[data_animes['Year'] >= 2000]

by_year = dat.sort_values('Year',ascending=True)
grupo=by_year.groupby("Year").mean()

fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Score", ax=ax, order=3, ci=None).set_title("Regresion")
plt.xticks(rotation=45)
plt.show()
````

````python
fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Watching", ax=ax, order=3, ci=None).set_title("Regresion")
plt.xticks(rotation=45)
plt.show()
````

````python
fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Completed", ax=ax, order=3, ci=None).set_title("Regresion")
plt.xticks(rotation=45)
plt.show()
````
